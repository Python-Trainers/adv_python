```python


import pandas as pd
import datetime
now = datetime.datetime.now()
nowstr =str(now.strftime("%m%d%Y%H%M%S"))
# Source Files locations...    
source_folder = r"/home/jovyan/adv_python/File Operations/"

from urllib.request import urlopen
from xml.etree.ElementTree import parse

var_url = urlopen('http://www.w3schools.com/xml/plant_catalog.xml')
xmldoc = parse(var_url)
columns = ['COMMON','BOTANICAL','ZONE,','LIGHT','PRICE','AVAILABILITY']
data_curr = pd.DataFrame(columns = columns)

for item in xmldoc.iterfind('PLANT'):
         c  = item.findtext('COMMON')
         b  = item.findtext('BOTANICAL')
         z  = item.findtext('ZONE')
         l  = item.findtext('LIGHT')
         p  = item.findtext('PRICE')
         a  = item.findtext('AVAILABILITY')        
         data_curr = data_curr.append(pd.Series([c,b,z,l,p,a], index = columns), ignore_index = True)        
print(data_curr)   

# Cerate the output text file...
data_curr.to_csv(source_folder+'XML_URL_oFile_'+nowstr+'.csv',index=False,sep=",",header=True)      
print("done")


```
