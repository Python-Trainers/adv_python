```python
# with basic logging

import logging

logging.debug('This is a debug message')
logging.info('This is an info message')
logging.warning('This is a warning message')
logging.error('This is an error message')
logging.critical('This is a critical message')
```

Debug - Details information, typically of interest only when diagnosing problems. 

Info - Confirmation that things are working as expected

Warning - An indication that something unexpected happened, or indicative of some problem in the near future (disk 
space low, etc). The software still working as expected

Error - Due to a more serious problem, the software has not been able to perform some functions

Critical - A serious error, indicating that the prgram itself may be unable to continue running.

Root - By Default you work with the root logger, specific loggers can be configured seperately

#### Log the error with logging.exception()

```python
import logging

import square_handler


#create a soecific logger for this file
logger = logging.getLogger(__name__)
#set the logLevel for the logger
logger.setLevel(logging.DEBUG)

#specift the formatting
formatter = logging.Formatter('%(asctime)s:%(levelname)s:%(message)s')

#specify the filename
file_handler = logging.FileHandler('app_new.log')
#set logging level to file handler
file_handler.setLevel(logging.ERROR)
#add the formatter to the file, so that the file also has formatt
file_handler.setFormatter(formatter)

#add the handler to the logger
logger.addHandler(file_handler)

#logging.basicConfig(filename = 'app.log', level=logging.DEBUG, format='%(asctime)s:%(levelname)s:%(message)s')

def add(num1, num2):
    return num1 + num2

def substract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    try:
        result = num1/num2
    except ZeroDivisionError as zde:
        #logger.exception('Invalid Division')
        logger.error('Invalid Division ', exc_info=True)
    else:
        logger.debug('Result = {}'.format(result))
        return result       
    

#Getting an error, so that the error will be recorded in the app_new.log file    
a=10
b=0

result_add = add(a, b)
logger.debug('Add : {} + {}  = {}'.format(a, b, result_add)) #new logger added


result_sub = substract(a, b)
logger.debug('Add : {} - {}  = {}'.format(a, b, result_sub))


result_mul = multiply(a, b)
logger.debug('Add : {} * {}  = {}'.format(a, b, result_mul))


result_div = divide(a, b)
logger.debug('Add : {} / {}  = {}'.format(a, b, result_div))
```

#### create a stream_handler

```python
import logging

import square_handler


#create a soecific logger for this file
logger = logging.getLogger(__name__)
#set the logLevel for the logger
logger.setLevel(logging.DEBUG)

#specift the formatting
formatter = logging.Formatter('%(asctime)s:%(levelname)s:%(message)s')

#specify the filename
file_handler = logging.FileHandler('app_new.log')
#set logging level to file handler
file_handler.setLevel(logging.ERROR)
#add the formatter to the file, so that the file also has formatt
file_handler.setFormatter(formatter)

#create a stream handler
stream_handler = logging.StreamHandler()
stream_handler.setFormatter(formatter)

#add the handler to the logger
logger.addHandler(file_handler)
logger.addHandler(stream_handler)

#logging.basicConfig(filename = 'app.log', level=logging.DEBUG, format='%(asctime)s:%(levelname)s:%(message)s')

def add(num1, num2):
    return num1 + num2

def substract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    try:
        result = num1/num2
    except ZeroDivisionError as zde:
        #logger.exception('Invalid Division')
        logger.error('Invalid Division ', exc_info=True)
    else:
        logger.debug('Result = {}'.format(result))
        return result       
    

#Getting an error, so that the error will be recorded in the app_new.log file    
a=10
b=0

result_add = add(a, b)
logger.debug('Add : {} + {}  = {}'.format(a, b, result_add)) #new logger added


result_sub = substract(a, b)
logger.debug('Add : {} - {}  = {}'.format(a, b, result_sub))


result_mul = multiply(a, b)
logger.debug('Add : {} * {}  = {}'.format(a, b, result_mul))


result_div = divide(a, b)
logger.debug('Add : {} / {}  = {}'.format(a, b, result_div))
```
